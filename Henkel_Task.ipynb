```python
df2= df.dropna(subset=['start station id', 'end station id','start station name','end station name'])
```

```python
df2['start station id'].isnull().values.any()
```

```python
df2['age'] = df2['birth year'].apply(lambda x: 2023-x)

df2
```

```python
subscribers= df2[df2['usertype'] == "Subscriber"]
subscribers.shape[0]
```

```python
customers= df2[df2['usertype'] == "Customer"]
customers.shape[0]
```

```python
usertype=['Subscriber','Customer']
count=[subscribers.shape[0],customers.shape[0]]

fig = plt.figure(figsize = (6, 6))
 
# creating the bar plot
plt.bar(usertype,count, color ='#7286D3',
        width = 0.4)

plt.xlabel("User Type")
plt.ylabel("Values")
plt.show()
```
